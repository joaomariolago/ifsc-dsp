## Ex 1

### Signal 1

$$
x_1(n) = 3\delta(n + 2) + 2\delta(n) - \delta(n - 3) + 5\delta(n - 7), \quad -5 < n < 15.
$$

```python
# Implementing: x_1(n) = 3\delta(n + 2) + 2\delta(n) - \delta(n - 3) + 5\delta(n - 7), \quad -5 < n < 15.

n = np.arange(-5, 15)
x = 3 * Signal.from_impulse(n, -2) + 2 * Signal.from_impulse(n) - Signal.from_impulse(n, 3) + 5 * Signal.from_impulse(n, 7)

x.stem(r'$x_1(n) = 3\delta(n + 2) + 2\delta(n) - \delta(n - 3) + 5\delta(n - 7), \quad -5 < n < 15$')
```

### Signal 2

$$
x_2(n) = \sum_{k=-5}^{5} e^{-|k|} \delta(n - 2k), \quad -10 \leq n \leq 10.
$$

```python
#Implementing x_2(n) = \sum_{k=-5}^{5} e^{-|k|} \delta(n - 2k), \quad -10 \leq n \leq 10.

n = np.arange(-10, 11)
x = Signal.from_zeros(n)

for k in range(-5, 6):
	x += np.exp(-np.abs(k)) * Signal.from_impulse(n, 2 * k)

x.stem(r'$x_2(n) = \sum_{k=-5}^{5} e^{-|k|} \delta(n - 2k), \quad -10 \leq n \leq 10$')
```

### Signal 3

$$
\quad x_3(n) = 10u(n) - 5u(n - 5) - 10u(n - 10) + 5u(n - 15). \\
$$

```python
# Implementing \quad x_3(n) = 10u(n) - 5u(n - 5) - 10u(n - 10) + 5u(n - 15)

n = np.arange(-5, 20)
x = 10 * Signal.from_step(n) - 5 * Signal.from_step(n, 5) - 10 * Signal.from_step(n, 10) + 5 * Signal.from_step(n, 15)

x.stem(r'$\quad x_3(n) = 10u(n) - 5u(n - 5) - 10u(n - 10) + 5u(n - 15)$')
```

### Signal 4

$$
\quad x_4(n) = e^{0.1n}[u(n + 20) - u(n - 10)].
$$

```python
# Implementing \quad x_4(n) = e^{0.1n}[u(n + 20) - u(n - 10)]

n = np.arange(-10, 21)
x = Signal.from_exponential(0.1, n) * (Signal.from_step(n, -20) - Signal.from_step(n, 10))

x.stem(r'$\quad x_4(n) = e^{0.1n}[u(n + 20) - u(n - 10)]$')
```

### Signal 5
$$
\quad x_5(n) = 5[\cos(0.49\pi n) + \cos(0.51\pi n)], \quad -200 \leq n \leq 200. \text{ Comment on the waveform shape.}
$$

```python
# Implementing \quad x_5(n) = 5[\cos(0.49\pi n) + \cos(0.51\pi n)], \quad -200 \leq n \leq 200

n = np.arange(-200, 201)
x = 5 * (Signal.from_cosine(0.49 * np.pi, n) + Signal.from_cosine(0.51 * np.pi, n))

x.stem(r'$\quad x_5(n) = 5[\cos(0.49\pi n) + \cos(0.51\pi n)], \quad -200 \leq n \leq 200$')
```

### Signal 6
$$
\quad x_6(n) = 2 \sin(0.01\pi n) \cos(0.5\pi n), \quad -200 \leq n \leq 200. \text{ Comment on the waveform shape.}

```python
# Implementing \quad x_6(n) = 2 \sin(0.01\pi n) \cos(0.5\pi n), \quad -200 \leq n \leq 200

n = np.arange(-200, 201)
x = 2 * Signal.from_sine(0.01 * np.pi, n) * Signal.from_cosine(0.5 * np.pi, n)

x.stem(r'$\quad x_6(n) = 2 \sin(0.01\pi n) \cos(0.5\pi n), \quad -200 \leq n \leq 200$')
```

### Signal 7

\quad x_7(n) = e^{-0.05n} \sin(0.1\pi n + \frac{\pi}{3}), \quad 0 \leq n \leq 100. \text{ Comment on the waveform shape.}

## Ex 2

### Signal 1

$$
\tilde{x}_1(n) = \{ \ldots, -2, -1, 0, 1, 2, \ldots \}_{\text{periodic}} \quad \text{Plot 5 periods.} \\
$$

```python
n = np.arange(0, 5)
x = Signal([-2, -1, 0, 1, 2], n)
x.stem(r'$\tilde{x}(n) = \{ -2, -1, 0, 1, 2 \}$')

x = x.tile(5)
x.stem(r'$\tilde{x}_1(n) = \{ \ldots, -2, -1, 0, 1, 2, \ldots \}_{periodic}$ Plot 5 periods.')
```

### Signal 2
$$
\tilde{x}_2(n) = e^{0.1\pi n} \left[u(n) - u(n - 20)\right]_{\text{periodic}} \quad \text{Plot 3 periods.}
$$

```python
n = np.arange(0, 20)
x = Signal.from_exponential(0.1 * np.pi, n) * (Signal.from_step(n) - Signal.from_step(n, 20))
x.stem(r'$\tilde{x}_2(n) = e^{0.1\pi n} \left[u(n) - u(n - 20)\right]$')

x = x.tile(3)
x.stem(r'$\tilde{x}_2(n) = e^{0.1\pi n} \left[u(n) - u(n - 20)\right]_{periodic}$ Plot 3 periods.')
```

### Signal 3

$$

\tilde{x}_3(n) = \sin(0.1\pi n) \left[u(n) - u(n - 10)\right]_{\text{periodic}} \quad \text{Plot 4 periods.} \\

$$

```python
n = np.arange(0, 10)

x = Signal.from_sine(0.1 * np.pi, n) * (Signal.from_step(n) - Signal.from_step(n, 10))
x.stem(r'$\tilde{x}_3(n) = \sin(0.1\pi n) \left[u(n) - u(n - 10)\right]$')

x = x.tile(4)
x.stem(r'$\tilde{x}_3(n) = \sin(0.1\pi n) \left[u(n) - u(n - 10)\right]_{periodic}$ Plot 4 periods.')
```

### Signal 4

$$
\tilde{x}_4(n) = \{ \ldots, 1, 2, 3, \ldots \}_{\text{periodic}} + \{ \ldots, 1, 2, 3, 4, \ldots \}_{\text{periodic}} \quad 0 \leq n \leq 24. \quad \text{What is the period of } \tilde{x}_4(n)?
$$

```python
n1 = np.arange(0, 3)
x1 = 1 + Signal.from_ramp(1, n1)

n2 = np.arange(0, 4)
x2 = 1 + Signal.from_ramp(1, n2)

x1.stem(r'$\tilde{x}_1(n) = \{ 1, 2, 3 \}$')
x2.stem(r'$\tilde{x}_1(n) = \{ 1, 2, 3, 4 \}$')

x1 = x1.tile(8)
x2 = x2.tile(6)

x3 = x1 + x2
x3.stem(r'$\tilde{x}_4(n) = \{ \ldots, 1, 2, 3, \ldots \}_{periodic} + \{ \ldots, 1, 2, 3, 4, \ldots \}_{periodic} \quad 0 \leq n \leq 24.$')
```

### Signal 1

$$
x_1(n) = 2x(n - 3) + 3x(n + 4) - x(n)
$$

```python
x1 = 2 * (x >> 3) + 3 * (x << 4) - x

x1.stem(r'$x_1(n) = 2x(n - 3) + 3x(n + 4) - x(n)$')
```

### Signal 2

$$
x_2(n) = 4x(4 + n) + 5x(n + 5) + 2x(n)
$$

```python
x2 = 4 * (x << 4) + 5 * (x << 5) + 2 * x

x2.stem(r'$x_2(n) = 4x(4 + n) + 5x(n + 5) + 2x(n)$')
```

### Signal 3

$$
x_3(n) = x(n + 3)x(n - 2) + x(1 - n)x(n + 1)
$$

```python
x3 = (x << 3) * (x >> 2) + (~x >> 1) * (x << 1)

x3.stem(r'$x_3(n) = x(n + 3)x(n - 2) + x(1 - n)x(n + 1)$')
```
